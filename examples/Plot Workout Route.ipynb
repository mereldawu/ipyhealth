```python
import random
import pandas as pd
from matplotlib import pyplot as plt

from ipyhealth import plotter
```

# Read routes dataframe

The routes.pkl dataframe will be the parsed output from AppleHealthFormatter.

If you have your own dataframe that constains latitude & lonitude columns then you can use it too.

```python
df = pd.read_pickle('data/routes.pkl')
filename = '/workout-routes/route_2020-02-16_2.46pm.gpx'
data = df[df.filename==filename]
```

# Generate heart rate

Heartrate is generated randomly here.

Heartrate is an optional param, if you don't have the heartrate info, you can leave it out.

```python
heartrate = [random.randint(70, 180) for i in data.latitude]
```

# Plot the Latitude & Lonitude with the corresponding to the Heartrate

```python
fig = plotter.plot_map(data, heartrate=heartrate)
plt.title('Hike on 16 February 2020');
```
